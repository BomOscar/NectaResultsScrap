```python
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas

url = "https://necta.go.tz/"

req = rq.get(url).content
rr = bs(req, "html.parser")

uls = rr.find_all("ul")
neww = []
for ul in uls:
    links = ul.find_all("a")
    for link in links:
        href = link.get("href")
        if href and href.endswith("_results"):
            full_link = url + href
            neww.append(full_link)
```

```python
for i in neww:
  res = rq.get(i).content
  captured = bs(res,"html").find("div", class_="main-content").find_all("ul")
  for i in captured:
    lnks  = i.find_all("a")
    for i in lnks:
      hr = i.get("href")
      if hr and hr.endswith(".htm"):
              print(hr)
















```
